# Testing Your Code
Тестирование вашего кода

Существует множество способов тестирования кода. В этом учебнике вы изучите приемы, начиная с самых простых шагов, и перейдете к продвинутым методам.

## Automated vs. Manual Testing
Автоматизированное и ручное тестирование

Хорошая новость заключается в том, что вы, скорее всего, уже создали тест, даже не подозревая об этом. Помните, когда вы запустили свое приложение и использовали его в первый раз? Вы проверяли функции и экспериментировали с их использованием? Это называется исследовательским тестированием (exploratory testing) и является одной из форм ручного тестирования (manual testing).

Исследовательское тестирование - это форма тестирования, которая проводится без плана. В исследовательском тестировании вы просто изучаете приложение.

Чтобы получить полный набор ручных тестов, достаточно составить список всех функций вашего приложения, различных типов входных данных, которые оно может принимать, и ожидаемых результатов. Теперь, каждый раз, когда вы вносите изменения в код, вам нужно пройтись по всем пунктам этого списка и проверить их.

Звучит не очень весело, не так ли?

Именно здесь на помощь приходит автоматизированное тестирование. Автоматизированное тестирование (Automated testing) - это выполнение вашего плана тестирования (части приложения, которые вы хотите протестировать, порядок, в котором вы хотите их протестировать, и ожидаемые ответы) скриптом вместо человека. Python уже поставляется с набором инструментов и библиотек, которые помогут вам создать автоматизированные тесты для вашего приложения. Мы рассмотрим эти инструменты и библиотеки в этом учебнике.

## Unit Tests vs. Integration Tests

Юнит тесты vs Интеграционные тесты

Мир тестирования не испытывает недостатка в терминологии, и теперь, когда вы знаете разницу между автоматизированным и ручным тестированием, пришло время углубиться на один уровень.

Подумайте, как можно проверить фары автомобиля. Вы включите фары (известный как этап/шаг тестирования)((known as the test step)) и выйдете из машины или попросите друга проверить, включены ли фары (известный как утверждение тестирования). Тестирование нескольких компонентов известно как интеграционное тестирование. (integration testing)

Подумайте обо всех вещах, которые должны работать правильно, чтобы простая задача дала нужный результат. Эти компоненты - как части вашего приложения, все те классы, функции и модули, которые вы написали.

Основная проблема интеграционного тестирования - это когда интеграционный тест не дает правильного результата. Очень трудно диагностировать проблему, не имея возможности определить, какая часть системы не работает. Если свет не включился, то, возможно, сломались лампочки. Может быть, сел аккумулятор? Что насчет генератора переменного тока? Компьютер автомобиля вышел из строя?

Если у вас модный современный автомобиль, он сообщит вам, когда перегорят лампочки. Для этого используется форма "модульного" теста. (unit test)

Юнит-тест - это более мелкий тест, который проверяет правильность работы отдельного компонента. Юнит-тест помогает вам изолировать то, что сломалось в вашем приложении, и быстрее исправить это.

Вы только что познакомились с двумя типами тестов:

* (An integration test) Интеграционный тест проверяет, что компоненты вашего приложения работают друг с другом.
* (A unit test) Юнит-тест проверяет работу небольшого компонента в вашем приложении.

В Python можно писать как Integration, так и Unit тесты. Чтобы написать модульный тест для встроенной функции sum(), нужно проверить вывод sum() на соответствие известному выводу.

Например, вот как проверить, что сумма() чисел (1, 2, 3) равна 6:

Оператор/Инструкция assert позволяет производить проверки истинности утверждений, что может быть использовано в отладочных целях.
Если проверка не прошла, вызывается исключение AssertionError.

https://docs.python.org/3/reference/simple_stmts.html#the-assert-statement



(assert, pass, return del are statements)

````python
assert sum([1, 2, 3]) == 6, "Should be 6"
````

Это ничего не выведет в REPL, потому что значения верны.

Если результат sum() неверен, это утверждение завершится с ошибкой AssertionError и сообщением "Should be 6". Попробуйте снова выполнить утверждение с неверными значениями, чтобы увидеть AssertionError:

````python
assert sum([1, 1, 1]) == 6, "Should be 6"
````

Вместо тестирования в REPL вы захотите поместить это в новый файл Python под названием test_sum.py и выполнить его снова:

````python
# test_sum.py
def test_sum():
    assert sum([1, 2, 3]) == 6, "Should be 6"

if __name__ == "__main__":
    test_sum()
    print("Everything passed")
````

Теперь вы написали тестовый пример, assert(рус. утверждение) и точку входа. 
Теперь вы можете выполнить скрипт в командной строке:
$ python test_sum.py
Everything passed

Вы можете видеть успешный результат, Everything passed.

В Python функция sum() принимает в качестве первого аргумента iterable object. Вы тестировали со списком. Теперь протестируйте с кортежем. Создайте новый файл test_sum_2.py со следующим кодом:

````python
# test_sum_2.py
def test_sum():
    assert sum([1, 2, 3]) == 6, "Should be 6"

def test_sum_tuple():
    assert sum((1, 2, 2)) == 6, "Should be 6"

if __name__ == "__main__":
    test_sum()
    test_sum_tuple()
    print("Everything passed")
````

Когда вы выполните test_sum_2.py, скрипт выдаст ошибку, потому что sum() из (1, 2, 2) равна 5, а не 6. В результате выполнения скрипта, вы получите следующий вывод:

$ python test_sum_2.py
Traceback (most recent call last):
  File "test_sum_2.py", line 9, in <module>
    test_sum_tuple()
  File "test_sum_2.py", line 5, in test_sum_tuple
    assert sum((1, 2, 2)) == 6, "Should be 6"
AssertionError: Should be 6

Здесь вы можете увидеть, как ошибка в вашем коде выдает ошибку на консоль с некоторой информацией о том, где была ошибка и каков был ожидаемый результат.

Примечание: С помощью doctest можно одновременно документировать и тестировать код, обеспечивая при этом синхронизацию кода и документации. Ознакомьтесь с doctest в Python: https://realpython.com/python-doctest/, чтобы узнать больше.

Написание тестов таким образом подходит для простой проверки, но что делать, если несколько тестов не работают? Здесь на помощь приходят программы для запуска тестов. Прогонщик тестов (The test runner) - это специальное приложение, предназначенное для запуска тестов, проверки результатов и предоставления вам инструментов для отладки и диагностики тестов и приложений.

Наиболее популярными являются три test runner'а:

### unittest

% python3 -m unittest --help

unittest встроен в стандартную библиотеку Python с версии 2.1. Вы, вероятно, увидите его в коммерческих приложениях Python и проектах с открытым исходным кодом.

unittest содержит как фреймворк для тестирования, так и программу запуска тестов. 
unittest имеет несколько важных требований для написания и выполнения тестов.

unittest требует, чтобы:

* Вы помещаете свои тесты в классы в виде методов
* Использовали серию специальных методов утверждения в классе unittest.TestCase вместо встроенного оператора assert.

Чтобы переписать предыдущий пример в тестовый пример unittest, необходимо:

1) Импортировать unittest из стандартной библиотеки
2) Создать класс TestSum, который наследуется от класса TestCase
3) Преобразовать тестовые функции в методы, добавив self в качестве первого аргумента.
4) Изменить утверждения для использования метода self.assertEqual() в классе TestCase.
5) Измените точку входа командной строки на вызов unittest.main().

Выполните эти шаги, создав новый файл test_sum_unittest.py со следующим кодом:

````python
# test_sum_unittest.py

import unittest


class TestSum(unittest.TestCase):

    def test_sum(self):
        self.assertEqual(sum([1, 2, 3]), 6, "Should be 6")

    def test_sum_tuple(self):
        self.assertEqual(sum((1, 2, 2)), 6, "Should be 6")

if __name__ == '__main__':
    unittest.main()
````

# Writing Your First Test 
Написание первого теста

Давайте соберем воедино все, чему вы научились до сих пор, и вместо тестирования встроенной функции sum() протестируем простую реализацию того же требования.

Создайте новую папку проекта и внутри нее создайте новую папку my_sum. Внутри my_sum создайте пустой файл __init__.py. Создание файла __init__.py означает, что папка my_sum может быть импортирована как модуль из родительского каталога.

Папка вашего проекта должна выглядеть следующим образом:
```
project/
│
└── my_sum/
    └── __init__.py
```

Откройте файл my_sum/__init__.py и создайте новую функцию sum(), которая принимает iterable object (список, кортеж или набор) и суммирует значения:

````python
def sum(arg):
    total = 0
    for val in arg:
        total += val
    return total
````

## Where to Write the Test
Где писать тест

Чтобы начать писать тесты, вы можете просто создать файл test.py, который будет содержать ваш первый тестовый пример. Поскольку файл должен быть способен импортировать ваше приложение, чтобы его можно было протестировать, вы хотите поместить test.py выше папки package, так что ваше дерево каталогов будет выглядеть примерно так:

```project/
│
├── my_sum/
│   └── __init__.py
|
└── test.py
```

Вы обнаружите, что по мере добавления все большего количества тестов ваш единственный файл станет беспорядочным и трудным для обслуживания, поэтому вы можете создать папку tests/ и разделить тесты на несколько файлов. Условно каждый файл должен начинаться с test_, так что все программы запуска тестов будут считать, что файл Python содержит тесты, которые нужно выполнить. Некоторые очень большие проекты разбивают тесты на большее количество подкаталогов в зависимости от их назначения или использования.

Примечание: Что делать, если ваше приложение представляет собой один скрипт?

Вы можете импортировать любые атрибуты скрипта, такие как классы, функции и переменные, используя встроенную функцию __import__(). 
Вместо ```from my_sum import sum``` вы можете написать следующее:
```
target = __import__("my_sum.py")
sum = target.sum
```
Преимущество использования ```__import__()``` в том, что вам не нужно превращать папку проекта в пакет, и вы можете указать имя файла. Это также полезно, если имя вашего файла пересекается с какими-либо пакетами стандартной библиотеки. Например, math.py столкнется с модулем math.

## How to Structure a Simple Test
Как структурировать простой тест

Прежде чем приступить к написанию тестов, вам необходимо принять несколько решений:

1) Что вы хотите тестировать?
2) Вы пишете юнит тест или интеграционный тест?

Тогда структура теста должна в общих чертах соответствовать этому рабочему процессу:

1) Создайте входные данные
2) Выполнить тестируемый код, фиксируя выходные данные
3) Сравнить результат с ожидаемым результатом


В данном приложении вы тестируете функцию sum(). В sum() можно проверить множество параметров поведения, например:

* Может ли она суммировать список целых чисел?
* Может ли она суммировать кортеж или множество?
* Может ли она суммировать список плавающих чисел?
* Что произойдет, если вы предоставите ей плохой атрибут, например, одно целое число или строку?
* Что произойдет, если одно из значений будет отрицательным?

Самым простым тестом будет список целых чисел. Создайте файл test.py со следующим кодом на языке Python:

````python
import unittest

from my_sum import sum


class TestSum(unittest.TestCase):
    def test_list_int(self):
        """
        Test that it can sum a list of integers
        """
        data = [1, 2, 3]
        result = sum(data)
        self.assertEqual(result, 6)

if __name__ == '__main__':
    unittest.main()
````

Этот пример кода:

1) Импортирует sum() из созданного вами пакета my_sum

2) Определяет новый класс TestSum, который наследуется от unittest.TestCase

3) Определяет метод теста .test_list_int() для проверки списка целых чисел. Метод .test_list_int():

* Объявит переменную data со списком чисел (1, 2, 3).
* Присвоит результат my_sum.sum(data) переменной result
* Убедится, что значение результата равно 6, используя метод .assertEqual() класса unittest.TestCase.
* Определяет точку входа в командную строку, которая запускает unittest test-runner .main()

## How to Write Assertions
Как писать утверждения

Последним шагом в написании теста является проверка вывода на соответствие известному ответу. Это известно как assertion (утверждение). Убедитесь, что тесты повторяемы, и проведите тест несколько раз, чтобы убедиться, что он каждый раз дает один и тот же результат.

# Executing Your First Test
Выполнение первого теста

Теперь, когда вы создали первый тест, вы хотите его выполнить. Конечно, вы знаете, что он пройдет, но прежде чем создавать более сложные тесты, вы должны убедиться, что можете успешно их выполнить.


## Executing Test Runners
Выполнение тест раннеров

Приложение Python, которое выполняет ваш тестовый код, проверяет утверждения и выдает результаты тестирования в консоль, называется программой запуска тестов. ( is called the test runner)

В нижней части test.py вы добавили этот небольшой фрагмент кода:
```
if __name__ == '__main__':
    unittest.main()
```

Это точка входа в командную строку. Это означает, что если вы выполните сценарий самостоятельно, запустив python test.py в командной строке, он вызовет unittest.main(). Это запустит программу запуска тестов, обнаружив все классы в этом файле, которые наследуются от unittest.TestCase.

Это один из многих способов выполнения программы запуска тестов unittest. Если у вас есть единственный файл с именем test.py, вызов python test.py - отличный способ начать работу.

Другой способ - использование командной строки unittest. Попробуйте это:

$ python -m unittest test

Это приведет к выполнению одного и того же тестового модуля (с именем test) через командную строку.

Вы можете указать дополнительные опции для изменения вывода. Одним из них является -v, означающий "verbose - подробно". Попробуйте сделать это:

$ python -m unittest -v test
test_list_int (test.TestSum) ... ok

----------------------------------------------------------------------
Ran 1 tests in 0.000s

Это выполнит один тест в файле test.py и выведет результаты на консоль. В режиме Verbose перечислены имена тестов, которые были выполнены первыми, а также результат каждого теста.

Вместо того чтобы указывать имя модуля, содержащего тесты, вы можете запросить автоматическое обнаружение, используя:

$ python -m unittest discover

Это позволит найти в текущем каталоге все файлы с именем test*.py и попытаться протестировать их.

Если у вас есть несколько тестовых файлов, при условии соблюдения шаблона именования test*.py, вы можете указать имя каталога вместо этого, используя флаг -s и имя каталога:

$ python -m unittest discover -s tests

unittest выполнит все тесты в едином плане тестирования и выдаст вам результаты.

Наконец, если ваш исходный код находится не в корне каталога, а в подкаталоге, например, в папке src/, вы можете указать unittest, где выполнять тесты, чтобы он мог правильно импортировать модули с помощью флага -t:

$ python -m unittest discover -s tests -t src

unittest перейдет в каталог src/, проверит все файлы test*.py в каталоге tests и выполнит их.

## Understanding Test Output
Понимание результатов тестирования

Это был очень простой пример, в котором все прошло успешно, поэтому сейчас вы попробуете провести неудачный тест и интерпретировать его результаты.

sum() должна быть способна принимать другие списки числовых типов, например, дроби.

В верхней части файла test.py добавьте оператор import, чтобы импортировать тип Fraction из модуля fractions стандартной библиотеки:
```
from fractions import Fraction
```

Теперь добавьте тест с утверждением, ожидающим неверное значение, в данном случае ожидая, что сумма 1/4, 1/4 и 2/5 будет равна 1:

````python
import unittest

from my_sum import sum


class TestSum(unittest.TestCase):
    def test_list_int(self):
        """
        Test that it can sum a list of integers
        """
        data = [1, 2, 3]
        result = sum(data)
        self.assertEqual(result, 6)

    def test_list_fraction(self):
        """
        Test that it can sum a list of fractions
        """
        data = [Fraction(1, 4), Fraction(1, 4), Fraction(2, 5)]
        result = sum(data)
        self.assertEqual(result, 1)

if __name__ == '__main__':
    unittest.main()
````

Если вы снова выполните тесты с помощью python -m unittest test, вы увидите следующий результат:

$ python -m unittest test
F.
======================================================================
FAIL: test_list_fraction (test.TestSum)
----------------------------------------------------------------------
Traceback (most recent call last):
  File "test.py", line 21, in test_list_fraction
    self.assertEqual(result, 1)
AssertionError: Fraction(9, 10) != 1

----------------------------------------------------------------------
Ran 2 tests in 0.001s

FAILED (failures=1)

В выводе вы увидите следующую информацию:

1) На первой строке показаны результаты выполнения всех тестов, один неудачный (F) и один пройденный (.).

2) Запись FAIL показывает некоторые подробности о неудачном тесте:

* Имя метода теста (test_list_fraction).
* Модуль теста (test) и тестовый пример (TestSum).
* обратный путь к неудачной строке. A traceback to the failing line
* Детали утверждения с ожидаемым результатом (1) и фактическим результатом (Fraction(9, 10)).

Помните, что вы можете добавить дополнительную информацию к выводу теста, добавив флаг -v к команде python -m unittest.
