```python
# ЗАДАНИЕ 2. Найти наибольшее и наименьшее значение функции на отрезке и отобразить точкой (на графике 1)
#Наименьшее и наибольшее значение функции на заданном отрезке

import numpy as np
import matplotlib.pyplot as plt
 
x = np.linspace(0.0, 5.0, 100)
y=7*np.log2(4*x)*np.cos(3*x-1)
# print (min(y))

indexminznach = np.argmin(y)
xMINimalnogo = x[indexminznach]
yMINimalnogo = y[indexminznach]
    
indexmaxznach = np.argmax(y)
# print(indexmaxznach)
xMAXimalnogo = x[indexmaxznach]
yMAXimalnogo = y[indexmaxznach]

plt.plot(x, y)
plt.plot(xMINimalnogo, yMINimalnogo, 'o', color = 'g')
plt.plot(xMAXimalnogo, yMAXimalnogo, 'o', color = 'r')


plt.show()
```

```python
# ЗАДАНИЕ 3. Составить уравнение касательной и нормали к точке, а затем построить их (на графике 1)

import numpy as np
import matplotlib.pyplot as plt
 
x = np.linspace(0.0, 5.0, 100)
y=7*np.log2(4*x)*np.cos(3*x-1)

# Пусть точкой М будет x0 под [37] индексом
x0 = x[37]

# Шаг a. Вычислим f(x0)
f_x0 = y[37]
# Шаг b. Найдём производную функции f(x)- нашли выше
# -21*log(4*x)*sin(3*x - 1)/log(2) + 7*cos(3*x - 1)/(x*log(2))
# Шаг c. Вычислим производную функции f(x) в точке x0
f_dariv_x0 = -21*log(4*x0)*sin(3*x0 - 1)/log(2) + 7*cos(3*x0 - 1)/(x0*log(2))
# Шаг d. Подставим x0 и вычисленные значения f(x0) и производную f(x0) в формулу для касательной
y_kasat = f_x0+f_dariv_x0*(x-x0)
# Шаг e. Подставим x0 и вычисленные значения f(x0) и производную f(x0) в формулу для нормали
y_normal = f_x0-(1/f_dariv_x0)*(x-x0)

plt.plot(x, y) 
plt.plot(x, y_kasat, ":m")
plt.plot(x, y_normal, "y--")

plt.show()
```

```python
# 4. Построить касательное расслоение (график 4) (то есть касательные к множеству точек кривой)

import numpy as np
import matplotlib.pyplot as plt
 
x = np.linspace(0.0, 5.0, 100)
y=7*np.log2(4*x)*np.cos(3*x-1)
plt.plot(x, y) 

for xi in x:
    
    index_xi = np.where(x == xi)[0] # нахожу индекс по значению переменной, чтобы найти значение функции от неё
    f_x0 = y[index_xi]
    f_dariv_x0 = -21*np.log(4*xi)*np.sin(3*xi - 1)/np.log(2) + 7*np.cos(3*xi - 1)/(xi*np.log(2))
    y_kasat = f_x0+f_dariv_x0*(x-xi)

    plt.plot(x, y_kasat, ":m")
    
plt.show()   
```
