```python
from time import time

''' Функция принимает на вход значения: n-количество столбцов и строк; data_fig-спискок координат уже стоящих фигур; 
    для уже поставленных фигур высчитывает клетки под её боем, проверяя их на выход за границы поля.
    Функция возвращает список из всех "оккупированных" клеток'''

def F_occupied_kletki(n: int, data_fig: list) -> list:
    occupied_kletki=[]
    for el in data_fig: #беру координаты первой поставленной фигуры и находу все действительные клетки, являющиеся полем её удара
        i=el[0]
        j=el[1]
        occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i-1,j-1],[i-2,j-2],[i-3,j-3],[i-4,j-4]])))
        occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i+1,j+1],[i+2,j+2],[i+3,j+3],[i+4,j+4]])))
        occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i-1,j+1],[i-2,j+2],[i-3,j+3],[i-4,j+4]])))
        occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i+1,j-1],[i+2,j-2],[i+3,j-3],[i+4,j-4]])))
    occupied_kletki.extend(data_fig)  
    
    return occupied_kletki

''' передаём функции значения: n-количество столбцов и строк; l-количество фигур, которые нужно поставить;
    data_fig-спискок координат уже стоящих фигур; occupied_kletki-список, куда будут добавляться "занятые" клетки;
    x, y - координаты предыдущей позиции; reshenie - переменная, которая сохранит последнее решение, чтобы вывести его на консоль
    Теперь, если l==0, т.е. все фигуры расставлены, мы созраняем расставновку в нужном формате в файл, иначе мы проходимся по воображаемому полю, созраняя текущую клетку для оптимизации.
    Теперь смотрим есть ли клетка в "оккупированных", елси есть, переходим к анализу следующей клетки, если нет, добавляем её в расставленные фигуры и высчитываем "оккупированные" клетки занаво.

    '''

def check(n: int,l: int, data_fig: list, occupied_kletki: list, f, x, y) -> None:
    global reshenie # переменная, которая сохранит последнее решение, чтобы вывести его на консоль
    if l==0:
        for i in range (len(data_fig)):
            data_fig[i]= tuple(data_fig[i])
            f.write (f" {data_fig[i]}")
        reshenie=data_fig
        f.write("\n")
        return False # чтобы можно было вернуть предыдущие значения переменным во избежании ошибочных результатов (закрытие рекурсии)
    else:
        # проходимся по воображаемому полю, учитывая предыдущие координаты, дабы избежать повторений
        for i in range (x,n+1): 
            for j in range (n+1): 
                if x==i and j<y:
                    continue
                else:
                    if ([i,j] in occupied_kletki):
                        pass
                    else:
                        data_fig = data_fig + [[i,j]]
                        occupied_kletki_do = occupied_kletki 
                        occupied_kletki = F_occupied_kletki (n, data_fig)
                        if not (check(n,l-1,data_fig,occupied_kletki,f,i,j)):
                            data_fig = data_fig[:-1]
                            occupied_kletki = occupied_kletki_do

reshenie= []
def main():
    m=time()
    t= open("input.txt", "r+")
    data = t.readlines()
    f = open("output.txt", "w")

    # убираю знаки табуляции, чтобы не мешали
    for i in range (len(data)):
        data[i] = data[i].strip()
    print(data)

    # преобразую список для удобства
    del0 = data[0].split() 
    n,l,k = int(del0[0]),int(del0[1]), int(del0[2]) # количество строк и столбцов,сколько фигур необходимо расставить,количество уже расставленных фигур
    n = n - 1
    del0=data.pop(0) # теперь в data только кооринаты начальных фигур

    # уберём пробел для удобства между координатами
    data_fig = [data[i].split() for i in range (k)]


    # изменим str на int
    for i in data_fig:
        i[0]=int(i[0])
        i[1]=int(i[1])
    
    # вызовем функцию
    occupied_kletki =[]
    occupied_kletki = F_occupied_kletki (n, data_fig) 

    check(n, l, data_fig, occupied_kletki,f,0,0)
    

    # печатаем на консоль
    if reshenie == []:
        f.write('No solutions')
        print('No solutions')
    else:
        occupied_kletki = F_occupied_kletki (n, reshenie)
        for i in range (len(occupied_kletki)):
            occupied_kletki[i]=tuple(occupied_kletki[i])
        n=n+1
        for i in range (n):
            x=''
            for j in range (n):
                if (i,j) in reshenie:
                    x+='#'
                elif (i,j) in occupied_kletki:
                    x+='*'
                else:
                    x+='0'
            print(' '.join(list(x)))

    # f.close()
    # t.close()
    print('Время: '+str(time()-m))

if __name__=="__main__":
    main()  
```

```python
from time import time

''' Функция принимает на вход значения: n-количество столбцов и строк; data_fig-спискок координат уже стоящих фигур
    для последней уже поставленной фигуры высчитывает клетки под её боем, проверяя их на выход за границы поля.
    Функция возвращает список из всех "оккупированных" клеток для данной фигуры'''

def F_occupied_kletki(n: int, data_fig: list) -> list:
    occupied_kletki = []
    i=data_fig[-1][0]
    j=data_fig[-1][1]
    occupied_kletki.extend([data_fig[-1]])
    occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i-1,j-1],[i-2,j-2],[i-3,j-3],[i-4,j-4]])))
    occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i+1,j+1],[i+2,j+2],[i+3,j+3],[i+4,j+4]])))
    occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i-1,j+1],[i-2,j+2],[i-3,j+3],[i-4,j+4]])))
    occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i+1,j-1],[i+2,j-2],[i+3,j-3],[i+4,j-4]])))
    return occupied_kletki

''' передаём функции значения: n-количество столбцов и строк; l-количество фигур, которые нужно поставить;
    data_fig-спискок координат уже стоящих фигур; occupied_kletki-список, куда будут добавляться "занятые" клетки;
    x, y - координаты предыдущей позиции; reshenie - переменная, которая сохранит последнее решение, чтобы вывести его на консоль
    Теперь, если l==0, т.е. все фигуры расставлены, мы созраняем расставновку в нужном формате в файл, иначе мы проходимся по воображаемому полю, созраняя текущую клетку для оптимизации.
    Теперь смотрим есть ли клетка в "оккупированных", елси есть, переходим к анализу следующей клетки, если нет, добавляем её в расставленные фигуры и высчитываем "оккупированные" клетки занаво.

    '''

def check(n: int,l: int, data_fig: list, occupied_kletki: list, f, x, y) -> None:
    global reshenie,occupied_kletki_rez # переменная, которая сохранит последнее решение, чтобы вывести его на консоль
    if l==0:
        for i in range (len(data_fig)):
            data_fig[i]= tuple(data_fig[i])
            f.write (f" {data_fig[i]}")
        reshenie=data_fig
        occupied_kletki_rez = occupied_kletki
        f.write("\n")
        return False # чтобы можно было вернуть предыдущие значения переменным во избежании ошибочных результатов (закрытие рекурсии)
    else:
        # проходимся по воображаемому полю, учитывая предыдущие координаты, дабы избежать повторений
        for i in range (x,n+1): 
            for j in range (n+1): 
                if x==i and j<y:
                    continue
                else:
                    if ([i,j] in occupied_kletki):
                        pass
                    else:
                        data_fig = data_fig + [[i,j]]
                        occupied_kletki_current = []
                        occupied_kletki_current.extend(occupied_kletki)
                        occupied_kletki_dop = F_occupied_kletki (n, data_fig)
                        occupied_kletki.extend(occupied_kletki_dop)
                        if not (check(n,l-1,data_fig,occupied_kletki,f,i,j)):
                            data_fig = data_fig[:-1]
                            occupied_kletki = occupied_kletki_current

reshenie= []
occupied_kletki_rez = []
def main():
    m=time()
    t= open("input.txt", "r+")
    data = t.readlines()
    f = open("output.txt", "w")

    # убираю знаки табуляции, чтобы не мешали
    for i in range (len(data)):
        data[i] = data[i].strip()

    # преобразую список для удобства
    del0 = data[0].split() 
    n,l,k = int(del0[0]),int(del0[1]), int(del0[2]) # количество строк и столбцов,сколько фигур необходимо расставить,количество уже расставленных фигур
    n = n - 1
    del0=data.pop(0) # теперь в data только кооринаты начальных фигур

    # уберём пробел для удобства между координатами
    data = [data[i].split() for i in range (k)]


    # изменим str на int
    for i in data:
        i[0]=int(i[0])
        i[1]=int(i[1])
    
    # вызовем функцию и для каждой уже поставленной фигуры найдём клетки боя
    occupied_kletki = []
    data_fig = []

    for i in range (len(data)):
        data_fig.append(data[i])
        occupied_kletki_dop = F_occupied_kletki (n, data_fig) 
        occupied_kletki.extend(occupied_kletki_dop)

    check(n, l, data_fig, occupied_kletki,f,0,0)
    

    # печатаем на консоль
    if reshenie == []:
        f.write('No solutions')
        print('No solutions')
    else:
        for i in range (len(occupied_kletki_rez)):
            occupied_kletki_rez[i]=tuple(occupied_kletki_rez[i])
        n=n+1
        for i in range (n):
            x=''
            for j in range (n):
                if (i,j) in reshenie:
                    x+='#'
                elif (i,j) in occupied_kletki_rez:
                    x+='*'
                else:
                    x+='0'
            print(' '.join(list(x)))

    # f.close()
    # t.close()
    print('Время: '+str(time()-m))

if __name__=="__main__":
    main()  
```
