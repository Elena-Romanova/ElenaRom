```python
import pygame as pg
import sys
from random import randint

WIDTH = 600
HEIGHT = 600
FPS = 60

class Basket(pg.sprite.Sprite):
    def __init__(self):
        # Вызовем конструктор родительского класса (Sprite)
        pg.sprite.Sprite.__init__(self)
        # Создаём с помощью клласса Surface поверхность для изображения 
        self.image: pg.Surface = pg.image.load('корзина.png').convert_alpha()
        self.image = pg.transform.scale(self.image, (170, 150)) # задаём новый размер
        # Создаём с помощью клласса Rect объект для взаимодействия с другими объектами
        self.rect: pg.Rect = self.image.get_rect(center = (WIDTH/2, HEIGHT- 80))

    def update(self):
        pass

class Bear(pg.sprite.Sprite):
    def __init__(self):
        # Вызовем конструктор родительского класса (Sprite)
        pg.sprite.Sprite.__init__(self)
        # Создаём с помощью клласса Surface поверхность для изображения 
        self.image: pg.Surface = pg.image.load('медведь.png').convert_alpha()
        self.image = pg.transform.scale(self.image, (100, 100)) # изменяем размер
        # Создаём с помощью клласса Rect объект для взаимодействия с другими объектами
        # при этом рандомим ширину, и указываем высоту, для плавного появления объекта
        self.rect: pg.Rect = self.image.get_rect(center = (randint(50, WIDTH-50), 0 - 50))

    def update(self):
        self.rect.y += 5
        pass

class Game:
    def __init__(self):
        self.game_score = 0
        self.visualize(self.game_score)

    @staticmethod
    def collideBalls(bears, basket, game_score):
        for bear in bears:
            if basket.rect.collidepoint(bear.rect.center):
                if bear.rect.center[1] == basket.rect.top:
                    game_score += 1
                    bear.kill()
        return game_score

    @staticmethod
    def visualize(game_score):
        # инициализируем pygame с помощью команды pygame.init(), которая необходима для использования команд модуля pygame.
        pg.init()
        pg.mixer.init()
        # создаём главную поверхность 
        window = pg.display.set_mode((600, 600))

        # создаём поверхность фон главной поверхности
        background = pg.image.load('фон.jpg').convert()
        # масштабируем фон по размеру главной поверхности
        background = pg.transform.smoothscale(background, window.get_size())
        # прикрепляем поверхность-фон к главной поверхности
        window.blit(background, (0, 0))
        clock = pg.time.Clock()

        # создаём группу для спрайтов и добавляем туда наши классы
        sprites = pg.sprite.Group()
        bear = Bear()
        basket = Basket()
        sprites.add(basket)
        sprites.add(bear)
        bears = []
        bears.append(bear)

        # создаём поверхность для отображения результатов game_score
        f = pg.font.SysFont('arial', 10)

        # обновляем поверхность
        pg.display.update()

        play = True
        new_bear = 0
        while play:
            ''' при первом входе в цикл, в sprites находится экземпляр класса Bear() и Basket() и 
                пока мы не добавим в другой экземпляр этого класса, при выполнении update() 
                будет выполняться update() прописанный в классе '''
            
            for i in pg.event.get():
                if i.type == pg.QUIT:
                    play = False
                    pg.quit() 
                    sys.exit()  
            
            if pg.mouse.get_focused(): # если курсор находится в пределах окна игры
                pos = pg.mouse.get_pos()
                # я создаю экземпляр класса Basket(), меняю его координаты в зависимости от координат курсора, и добаляю его в sprites
                # basket = Basket()
                print(basket.rect.top)
                basket.rect.center = (pos[0], HEIGHT-80)
                # sprites.add(basket)

            # благодаря этому условию, медведи идут постепенно
            new_bear += 1
            if new_bear == 30:
                bear = Bear()
                bears.append(bear)
                sprites.add(bear)
                new_bear = 0
            clock.tick(FPS)    

            # добавляем медведя и корзинку на поверхность, и обновляем фон
            sprites.update()
            window.blit(background, (0, 0))
            sprites.draw(window)

            text = f.render(f"{game_score}", False, (255, 215, 0))
            window.blit(text, (3,590))

            # обновляем всю поверхность после того, как всё было отрисовано            
            pg.display.flip()
            clock.tick(FPS)
            


            game_score = Game.collideBalls(bears, basket, game_score)
            print(game_score)


Game()
```
