**Задание 1:**
Добавить в Листинг 2: деление, умножение. 

```python
from tkinter import Tk, Label, Button, Entry, IntVar, END, W, E
from tkinter.messagebox import showerror
class Calculator:

    def __init__(self, master:'Tk'):
        self.master = master
        master.title("Calculator")
        self.total = 0
        self.entered_number = 0
        self.total_label_text = IntVar()
        self.total_label_text.set(self.total)
        self.total_label = Label(master, textvariable=self.total_label_text)
        self.label = Label(master, text="Total:")
        vcmd = master.register(self.validate)
        self.entry = Entry(master, validate="key", validatecommand=(vcmd, '%P'))
        self.add_button = Button(master, text="+", command=lambda: self.update("add"))
        self.subtract_button = Button(master, text="-", command=lambda: self.update("subtract"))
        self.reset_button = Button(master, text="Reset", command=lambda: self.update("reset"))

        # добавляем умножение и деление
        self.multiplication_button = Button(master, text="*", command=lambda: self.update("multiplication"))
        self.division_button = Button(master, text="/", command=lambda: self.update("division"))

        self.label.grid(row=0, column=0, sticky=W)
        self.total_label.grid(row=0, column=1, columnspan=2, sticky=E)
        self.entry.grid(row=1, column=0, columnspan=3, sticky=W+E)
        self.add_button.grid(row=2, column=0)
        self.subtract_button.grid(row=2, column=1)

        # добавим кнопки умножени и деления
        self.multiplication_button.grid(row=2, column=2)
        self.division_button.grid(row=2, column=3)

        self.reset_button.grid(row=2, column=4, sticky=W+E)


    def validate(self, new_text: str):
        if not new_text: # the field is being cleared
            self.entered_number = 0
            return True
        try:
            self.entered_number = int(new_text)
            return True
        except ValueError:
            return False
    def update(self, method):
        if method == "add":
            self.total += self.entered_number
        elif method == "subtract":
            self.total -= self.entered_number
        # добавляем умножение
        elif method == "multiplication":
            self.total *= self.entered_number
        # добавляем деление и ошибку при делении на ноль
        elif method == "division":
            if self.entered_number == 0:
                showerror(title="Ошибка", message="На 0 делить нельзя")
            else:
                self.total /= self.entered_number
        else: # reset
            self.total = 0
        self.total_label_text.set(self.total)
        self.entry.delete(0, END)

root = Tk()
my_gui = Calculator(root)
root.mainloop()
```

**Задание 2:**
Обработка динамического списка. В главном окне 8 кнопок:

1) Создание списка 
2) Вывод списка в консоль
3) Запись списка в файл
4) Количество элементов в списке. 
5) Добавление элемента в список 
6) Поиск элемента в списке 
7) Удаление элемента из списка 
8) Выход 

Каждая кнопка открывает новое окно для следующих действий: 1, 4, 5, 6, 7.

```python
from tkinter import Tk, Label, Button, Entry, IntVar, END, W, E
from tkinter.messagebox import showinfo, showerror

class List_handler:
    def __init__(self, master) -> None:
        self.master = master
        master.title("Обработчик динамического списка")
        self.screenwidth = master.winfo_screenwidth()//2 
        self.screenheight = master.winfo_screenheight()//2
        master.geometry(f"270x210+{self.screenwidth-135}+{self.screenheight-105}")

        self.add_list = Button(master, text="Создание списка", command = self.add_list)
        self.output_list_to_console = Button(master, text="Вывод списка в консоль", command = self.output_list_to_console)
        self.writing_list_to_file = Button(master, text="Запись списка в файл", command=self.writing_list_to_file)
        self.number_of_elements_in_list = Button(master, text="Количество элементов в списке")
        self.add_elements_in_list = Button(master, text="Добавление элемента в список")
        self.search_elements_in_list = Button(master, text="Поиск элемента в списке")
        self.remove_elements_in_list = Button(master, text="Удаление элемента из списка")
        self.exit = Button(master, text="Выход ")

        self.add_list.pack(fill="x")
        self.output_list_to_console.pack(fill="x")
        self.writing_list_to_file.pack(fill="x")
        self.number_of_elements_in_list.pack(fill="x")
        self.add_elements_in_list.pack(fill="x")
        self.search_elements_in_list.pack(fill="x")
        self.remove_elements_in_list.pack(fill="x")
        self.exit.pack(fill="x")

    def add_list(self):
        self.new_list = [1,1,1]
        if hasattr(self, 'new_list'):
            showinfo(title="Уведомление", message="Список был создан")

    def output_list_to_console(self):
        if hasattr(self, 'new_list'):
            print(self.new_list)
            showinfo(title="Уведомление", message="Вывод списка в консоль выпонен")
        else:
            showerror(title="Ошибка", message="Список не был создан")

    def writing_list_to_file(self):
        f = open("output_task2.txt", "r") 
        # f.write(str(self.new_list) +'\n')
        print(f.read())
        showinfo(title="Уведомление", message="Запись списка в файл выпонена")


    

window = Tk()
play = List_handler(window) 
window.mainloop()
```
