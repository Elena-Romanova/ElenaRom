**Задание 1:**
Добавить в Листинг 2: деление, умножение. 

```python
from tkinter import Tk, Label, Button, Entry, IntVar, END, W, E
from tkinter.messagebox import showerror
class Calculator:

    def __init__(self, master:'Tk'):
        self.master = master
        master.title("Calculator")
        self.total = 0
        self.entered_number = 0
        self.total_label_text = IntVar()
        self.total_label_text.set(self.total)
        self.total_label = Label(master, textvariable=self.total_label_text)
        self.label = Label(master, text="Total:")
        vcmd = master.register(self.validate)
        self.entry = Entry(master, validate="key", validatecommand=(vcmd, '%P'))
        self.add_button = Button(master, text="+", command=lambda: self.update("add"))
        self.subtract_button = Button(master, text="-", command=lambda: self.update("subtract"))
        self.reset_button = Button(master, text="Reset", command=lambda: self.update("reset"))

        # добавляем умножение и деление
        self.multiplication_button = Button(master, text="*", command=lambda: self.update("multiplication"))
        self.division_button = Button(master, text="/", command=lambda: self.update("division"))

        self.label.grid(row=0, column=0, sticky=W)
        self.total_label.grid(row=0, column=1, columnspan=2, sticky=E)
        self.entry.grid(row=1, column=0, columnspan=3, sticky=W+E)
        self.add_button.grid(row=2, column=0)
        self.subtract_button.grid(row=2, column=1)

        # добавим кнопки умножени и деления
        self.multiplication_button.grid(row=2, column=2)
        self.division_button.grid(row=2, column=3)

        self.reset_button.grid(row=2, column=4, sticky=W+E)


    def validate(self, new_text: str):
        if not new_text: # the field is being cleared
            self.entered_number = 0
            return True
        try:
            self.entered_number = int(new_text)
            return True
        except ValueError:
            return False
    def update(self, method):
        if method == "add":
            self.total += self.entered_number
        elif method == "subtract":
            self.total -= self.entered_number
        # добавляем умножение
        elif method == "multiplication":
            self.total *= self.entered_number
        # добавляем деление и ошибку при делении на ноль
        elif method == "division":
            if self.entered_number == 0:
                showerror(title="Ошибка", message="На 0 делить нельзя")
            else:
                self.total /= self.entered_number
        else: # reset
            self.total = 0
        self.total_label_text.set(self.total)
        self.entry.delete(0, END)

root = Tk()
my_gui = Calculator(root)
root.mainloop()
```

**Задание 2:**
Обработка динамического списка. В главном окне 8 кнопок:

1) Создание списка 
2) Вывод списка в консоль
3) Запись списка в файл
4) Количество элементов в списке. 
5) Добавление элемента в список 
6) Поиск элемента в списке 
7) Удаление элемента из списка 
8) Выход 

Каждая кнопка открывает новое окно для следующих действий: 1, 4, 5, 6, 7.

```python
from tkinter import Tk, Label, Button, Entry, IntVar, END, W, E
from tkinter.messagebox import showinfo, showerror

class List_handler:
    def __init__(self, master) -> None:
        self.master = master
        master.title("Обработчик динамического списка")
        self.screenwidth = master.winfo_screenwidth()//2 
        self.screenheight = master.winfo_screenheight()//2
        master.geometry(f"270x210+{self.screenwidth-135}+{self.screenheight-105}")

        self.add_list = Button(master, text = "Создание списка", command = self.add_list)
        self.add_list.pack(fill="x")
        self.output_list_to_console = Button(master, text = "Вывод списка в консоль", command = self.output_list_to_console)
        self.output_list_to_console.pack(fill="x")
        self.writing_list_to_file = Button(master, text = "Запись списка в файл", command=self.writing_list_to_file)
        self.writing_list_to_file.pack(fill="x")
        self.number_of_elements_in_list = Button(master, text = "Количество элементов в списке", command = self.number_of_elements_in_list)
        self.number_of_elements_in_list.pack(fill="x")
        self.add_elements_in_list = Button(master, text = "Добавление элемента в список", command=self.add_elements_in_list)
        self.add_elements_in_list.pack(fill="x")
        self.search_elements_in_list = Button(master, text = "Поиск элемента в списке", command=self.search_elements_in_list)
        self.search_elements_in_list.pack(fill="x")
        self.remove_elements_in_list = Button(master, text = "Удаление элемента из списка", command=self.remove_elements_in_list)
        self.remove_elements_in_list.pack(fill="x")
        self.exiti = Button(master, text = "Выход ", command=self.exiti)
        self.exiti.pack(fill="x")

    def exiti(self):
        self.master.destroy()

    def add_list(self):
        self.window_dop=Tk()
        self.window_dop.geometry(f"+{self.screenwidth-135}+{self.screenheight-105}")
        self.new_list = []
        self.label = Label(self.window_dop,text='Введите элемент(ы)')
        self.entry = Entry(self.window_dop, width=25)
        self.button = Button(self.window_dop,text='Добавить',command=self.add_elements)
        self.label.grid(row = 1, column=0)
        self.entry.grid(row = 1, column=1)
        self.button.grid(row = 1, column=2,padx=4)
        self.prim = Label(self.window_dop,text='* элементы должны быть целыми числами, и записываться через ","')
        self.prim.grid(row = 0, columnspan=3, pady=2)
    def add_elements(self):
        dell = self.entry.get().split(',')
        for i in dell:
            self.new_list.append(int(i))
        self.window_dop.destroy()
        showinfo(title="Уведомление", message=f"Список {self.new_list} был успешно создан/обновлён")

    def output_list_to_console(self):
        if hasattr(self, 'new_list'):
            print(self.new_list)
            showinfo(title="Уведомление", message="Вывод списка в консоль выпонен")
        else:
            showerror(title="Ошибка", message="Список не был создан")

    def writing_list_to_file(self):
        f = open("output_task2.txt", "r+") 
        print(f.read())
        f.write(str(self.new_list) +'\n')
        showinfo(title="Уведомление", message="Запись списка в файл выпонена")

    def number_of_elements_in_list(self):
        showinfo(title="Уведомление", message=f'В списке {len(self.new_list)} элементов')
    
    def add_elements_in_list(self):
        self.window_dop=Tk()
        self.window_dop.geometry(f"+{self.screenwidth-135}+{self.screenheight-105}")
        self.label = Label(self.window_dop,text='Введите элемент(ы)')
        self.entry = Entry(self.window_dop, width=25)
        self.button = Button(self.window_dop,text='Добавить',command=self.add_elements)
        self.label.grid(row = 1, column=0)
        self.entry.grid(row = 1, column=1)
        self.button.grid(row = 1, column=2,padx=4)

    def search_elements_in_list(self):
        self.window_dop=Tk()
        self.window_dop.geometry(f"+{self.screenwidth-135}+{self.screenheight-105}")
        self.entry=Entry(self.window_dop,width=25)
        self.label = Label(self.window_dop,text='Введите элемент для поиска')
        self.label.pack(side='left')
        self.entry.pack(side='left')
        self.button=Button(self.window_dop,text='Найти',command=self.search_elements)
        self.button.pack(side='left')
    def search_elements(self):
        dell = int(self.entry.get())
        if dell in self.new_list:
            if self.new_list.count(dell) == 1:
                showinfo(title="Уведомление", message=f"Элемент найден в списке, его индекс {self.new_list.index(dell)}")
            else:
                showinfo(title="Уведомление", message=f"Элемент найден в списке, он встерчается {self.new_list.count(dell)} раз")
        else:
            showinfo(title="Уведомление", message="Элемент в списке отсутсвует")
        self.window_dop.destroy()

    def remove_elements_in_list(self):
        self.window_dop=Tk()
        self.window_dop.geometry(f"+{self.screenwidth-135}+{self.screenheight-105}")
        Label(self.window_dop,text='Введите элемент для удаления').pack(side='left')
        self.entry=Entry(self.window_dop,width=25)
        self.entry.pack(side='left')
        Button(self.window_dop,text='Удалить',command=self.deleted).pack(side='left')
    def deleted(self):
        dell=int(self.entry.get())
        if dell in self.new_list:
            self.new_list.remove(dell)
            showinfo(title="Уведомление", message=f"Элемент {dell} удалён")
        else:
            showerror(title="Ошибка", message="Элемент в списке отсутсвует")
        self.window_dop.destroy()


window = Tk()
play = List_handler(window) 
window.mainloop()
```

**Задание 3:**

Сделать меню, со списком фигур. 

Данная фигура является классом (Взять три любые фигуры из задания, которое давал ранее. Иерархия классов должна сохранится).

Вам необходимо их нарисовать и вывести площадь на экран. 

Ввод данных производится в окне. 

Можете добавить кнопку.

```python
from tkinter import *
import math

class Menu:
    def __init__(self, master) -> None:
        self.master = master
        master.title("Нарисует и выведет площадь")
        self.screenwidth = master.winfo_screenwidth()//2 
        self.screenheight = master.winfo_screenheight()//2
        master.geometry(f"+{self.screenwidth-135}+{self.screenheight-105}")

        self.label = Label(self.master, text = 'Выберите желаемую фигуру:')
        self.label.pack(fill='x', side='left', padx=3, pady=2)
        self.label_square = Button(self.master, text = 'Квадрат', command= self.square)
        self.label_square.pack(fill='x', side='left', padx=3, pady=2)
        self.label_circle = Button(self.master, text = 'Круг', command= self.circle)
        self.label_circle.pack(fill='x', side='left', padx=3, pady=2)
        self.label_rectangle = Button(self.master, text = 'Прямоугольник', command=self.rectangle)
        self.label_rectangle.pack(fill='x', side='left', padx=3, pady=2)

    def square(self):
        square = Square()
        square.draw_square()

    def circle(self):
        circle = Circle()
        circle.draw_circle()

    def rectangle(self):
        rectangle = Rectangle()
        rectangle.draw_rectangle()

class Square:
    def __init__(self) -> None:
        self.window = Tk()
        self.window.geometry("+600+300")

    def draw(self, x):
        if hasattr(self, 'canvas'):
            self.canvas.destroy()
            self.label.destroy()

        self.label = Label(self.window, text = f'Площадь равна {x**2}')
        self.label.pack()
        self.canvas = Canvas(self.window, bg="white", width=250, height=250)
        self.canvas.pack(anchor=CENTER)
        self.canvas.create_rectangle(125-x//2, 125-x//2, 125+x//2, 125+x//2, fill="yellow", outline="#004D40")

    def draw_square(self):
        self.label = Label(self.window, text = 'Введите длину стороны (<250):')
        self.label.pack()
        self.entry = Entry(self.window, width=30)
        self.entry.pack()
        self.button = Button(self.window, text='Нарисовать', command = lambda: self.draw(int(self.entry.get())))
        self.button.pack()

        

class Circle:
    def __init__(self) -> None:
        self.window = Tk()
        self.window.geometry("+600+300")

    def draw(self, x):
        if hasattr(self, 'canvas'):
            self.canvas.destroy()
            self.label.destroy()

        self.label = Label(self.window, text = f'Площадь равна {math.pi*x**2}')
        self.label.pack()    
        self.canvas = Canvas(self.window, bg="white", width=250, height=250)
        self.canvas.pack(anchor=CENTER)
        self.canvas.create_oval(125-x//2, 125-x//2, 125+x//2, 125+x//2, fill="yellow", outline="#004D40")

    def draw_circle(self):
        self.label = Label(self.window, text = 'Введите радиус (<200):')
        self.label.pack()
        self.entry = Entry(self.window, width=30)
        self.entry.pack()
        self.button = Button(self.window, text='Нарисовать', command = lambda: self.draw(int(self.entry.get())))
        self.button.pack()  

class Rectangle:
    def __init__(self) -> None:
        self.window = Tk()
        self.window.geometry("+600+300")

    def draw(self, x, y):
        if hasattr(self, 'canvas'):
            self.canvas.destroy()
            self.label.destroy()

        self.label = Label(self.window, text = f'Площадь равна {x*y}')
        self.label.pack()
        self.canvas = Canvas(self.window, bg="white", width=250, height=250)
        self.canvas.pack(anchor=CENTER)
        self.canvas.create_rectangle(125-x//2, 125-y//2, 125+x//2, 125+y//2, fill="yellow", outline="#004D40")

    def draw_rectangle(self):
        self.lbl = Label(self.window, text = 'Введите длины сторон:')
        self.lbl.pack()
        self.entry_1 = Entry(self.window, width=30)
        self.entry_1.pack()
        self.entry_2 = Entry(self.window, width=30)
        self.entry_2.pack()
        self.button = Button(self.window, text='Нарисовать', command = lambda: self.draw(int(self.entry_1.get()), int(self.entry_2.get())))
        self.button.pack()

window = Tk()
menu = Menu(window)
window.mainloop() 
```

**Задание 4:**
1) Напишите программу, которая случайным образом выбирает целое число от 1 до 100. Ваша программа должна запрашивать у пользователя догадки — если пользователь угадывает неправильно, она должна печатать, является ли догадка слишком высокой или слишком низкой. Если пользователь угадывает правильно, программа должна вывести количество догадок, сделанных пользователем, чтобы угадать правильный ответ. Вы можете предположить, что пользователь введет правильный ввод.
2) Вы должны создать класс, который будет принимать экземпляр класса Tk().
3) Должны быть три кнопки:
·      Ввод значения 
·      Начать заново 
·      Вывод сообщения о количестве попыток
4) Должна быть надпись о догадках куда следовать (верх, вниз, правильный ответ)
5) Значение выбирается рандомно
6) Ввод строки, где будет проверяется правильность введенного значения (тут нужно быть внимательнее и проверять, на то чтобы буквы не вводились и  небыло выхода за границы поиска) 

```python
from tkinter import * 
from random import randint

class Guess:
    def __init__(self, master):
        self.master = master
        master.title("Зарандомит число, попробуй угадай")
        self.screenwidth = master.winfo_screenwidth()//2 
        self.screenheight = master.winfo_screenheight()//2
        master.geometry(f"+{self.screenwidth-135}+{self.screenheight-105}")
        
        self.count = 0
        self.number = randint(1,101)

        self.entry = Entry(self.master, width=20)
        self.entry.grid(row=0, column=0)
        self.button_1 = Button(self.master,text = "Проверка значения",width=20, command= lambda: self.enter_value())
        self.button_1.grid(row=0, column=1)
        self.button_2 = Button(self.master,text = "Попробовать снова",width=20, command= lambda: self.again())
        self.button_2.grid(row=1, column=0)
        self.button_3 = Button(self.master,text = "Количество попыток",width=20, command= lambda: self.number_of_attempts())
        self.button_3.grid(row=1, column=1)

        self.label = Label(self.master, text = '')
        self.label.grid(row=2, columnspan=2)

    def number_of_attempts(self):
        showinfo(title="Уведомление", message=f'Количество попыток: {self.count} \n') 

    def again(self):
        self.label.config(text = "")
        self.count = 0
        self.number = randint(1,101)

    def enter_value(self):
        self.popitka = self.entry.get()

        self.count+=1
        if int(self.popitka) > 100 or int(self.popitka) <0:
            self.label.config(text = "Выход за пределы")
        elif int(self.popitka) == int(self.number):
            self.label.config(text ="Правильный ответ" )
        elif int(self.popitka) > int(self.number):
            self.label.config(text = "Меньше")
        elif int(self.popitka) < int(self.number):
            self.label.config(text = "Больше")

window = Tk()
guess = Guess(window)
window.mainloop()
```
