# Вернемся к принципам ООП



Четыре основных принципа объектно-ориентированного программирования следующие.
* **Абстракция**. Моделирование требуемых атрибутов и взаимодействий сущностей в виде классов для определения абстрактного представления системы.
* **Инкапсуляция**. Скрытие внутреннего состояния и функций объекта и предоставление доступа только через открытый набор функций.
* **Наследование**. Возможность создания новых абстракций на основе существующих.
* **Полиморфизм**. Возможность реализации наследуемых свойств или методов отличающимися способами в рамках множества абстракций.

Abstraction, Encapsulation, Inheritance, Polymorphism

Также существуют такие принципы ооп, как Ассоциация (Агрегация, Композиция).Это отношения между двумя классами, и эти отношения устанавливаются через их объекты. Каждый объект имеет свой жизненный цикл, и нет объекта-владельца. Это слабый тип отношений. Это может быть один к одному, один ко многим, многие к одному или многие ко многим.


**Агрегация** — это концепция, согласно которой объект одного класса может владеть или получать доступ к другому независимому объекту другого класса. 

- Он представляет отношения Has-A. # In the Car class: self.engine = engine # Has-A Engine
- Это однонаправленная ассоциация , т.е. односторонняя связь. Например, на факультете могут быть студенты, но наоборот, это невозможно и, следовательно, имеет однонаправленный характер.
- В агрегации обе записи могут сохраняться по отдельности, что означает, что завершение одной сущности не повлияет на другую сущность.




В UML это обозначается прямой линией с пустой стрелкой на одном конце.



````python
class Student:

    def __init__(self, id):
        self._id = id

    def registration_number(self, department_id) -> str:
        return str(self._id) + '-' + department_id


class Department:

    def __init__(self, id, student):
        self._id = id
        self._student = student

    def student_registration(self):
        return self._student.registration_number(self._id)


if __name__ == '__main__':
    student = Student(1)
    department = Department('ENG', student)
    print(department.student_registration())
````

Агрегация почти полностью аналогична композиции. Разница в том, что агрегированные объекты могут существовать независимо друг от друга.

**Композиция** —это концепция, согласно которой два объекта чрезвычайно зависят друг от друга.

- Он представляет отношения Has-A.
- Он указывает на компонент отношений.
- Обе сущности зависят друг от друга по составу.
- Составной объект не может существовать без другого объекта, когда между двумя объектами существует композиция.

В UML это обозначается прямой линией с закрашенной стрелкой на одном конце.



Это такой тип отношений, когда класс именно «владеет» объектом и несёт ответственность за его время жизни

````python
class Student:

    def __init__(self, id):
        self._id = id

    def registration_number(self, department_id) -> str:
        return str(self._id) + '-' + department_id


class Department:

    def __init__(self, department_id, student_id):
        self._id = department_id
        self._student = Student(student_id)

    def student_registration(self):
        return self._student.registration_number(self._id)


if __name__ == '__main__':
    department = Department('ENG', 10)
    print(department.student_registration())
````

Также следует помнить, что композиция - это агрегация; агрегация - это просто более общая форма композиции. Любое композиционное отношение также является агрегационным отношением, но не наоборот.

## Абстракция

Моделирование требуемых атрибутов и взаимодействий сущностей в виде классов для определения абстрактного представления системы.

Абстракция - термин ооп, связанный с инкапсуляцией и сокрытием информации.

Абстракция - одна из важнейших особенностей объектно-ориентированного программирования. Она используется для того, чтобы скрыть фоновые детали или любую ненужную реализацию.

Предопределенные функции похожи на абстракцию данных.
Например, когда вы используете стиральную машину для стирки белья. Вы кладете белье и стиральный порошок в машину и ждете, пока машина выполнит свою задачу. Как она ее выполняет? Какой механизм она использует? Пользователю не нужно знать, какой механизм лежит в основе ее работы. Этот процесс обычно известен как абстрагирование данных, когда вся ненужная информация скрывается от пользователей.

В Python мы можем достичь абстракции путем включения абстрактных классов и методов.

Любой класс, содержащий абстрактный метод(ы), называется абстрактным классом. Абстрактные методы не содержат реализаций - они всегда определяются и реализуются как часть методов подклассов, унаследованных от абстрактного класса. Посмотрите на приведенный ниже пример синтаксиса абстрактного класса:
```
from abc import ABC
// abc is a library from where a class ABC is being imported. However, a separate class can also be created. The importing from the library has nothing to do with abstraction.

Class type_shape(ABC):
```

Класс type_shape наследуется от класса ABC. Определим абстрактный метод area внутри класса type_shape:

```
from abc import ABC,abstractmethod
class type_shape(ABC):
  // abstract method area
  @abstractmethod
  def area(self):
    pass
```

Реализация абстрактного класса осуществляется в подклассах, которые наследуют класс type_shape. В приведенном ниже коде мы определили четыре класса, которые наследуют абстрактный класс type_shape:

````python
from abc import ABC, abstractmethod
class type_shape(ABC): 
  @abstractmethod
  def area(self): 
    #abstract method
    pass
  

  # @classmethod
  # @abstractmethod
  # def func(cls):
  #   pass
  
  # @staticmethod
  # @abstractmethod
  # def func2():
  #   pass
  
  # @property
  # @abstractmethod
  # def func3(self):
  #   pass

class Rectangle(type_shape):
  length = 6
  breadth = 4
  def area(self):
    return self.length * self.breadth

class Circle(type_shape):
  radius = 7
  def area(self):
    return 3.14 * self.radius * self.radius

class Square(type_shape):
  length = 4
  def area(self):
    return self.length*self.length

class triangle:
  length = 5
  width = 4
  def area(self):
    return 0.5 * self.length * self.width


r = Rectangle() # object created for the class 'Rectangle'
c = Circle() # object created for the class 'Circle'
s = Square() # object created for the class 'Square'
t = triangle() # object created for the class 'triangle'
print("Area of a rectangle:", r.area()) # call to 'area' method defined inside the class.
print("Area of a circle:", c.area()) # call to 'area' method defined inside the class.
print("Area of a square:", s.area()) # call to 'area' method defined inside the class.
print("Area of a triangle:", t.area()) # call to 'area' method defined inside the class.
````

## Инкапсуляция

Скрытие внутреннего состояния и функций объекта и предоставление доступа только через открытый набор функций.

Она описывает идею обертывания данных и методов, которые работают с данными, в единое целое. Это накладывает ограничения на прямой доступ к переменным и методам и может предотвратить случайное изменение данных. Чтобы предотвратить случайное изменение, переменная объекта может быть изменена только методом объекта. Такие типы переменных известны как приватные переменные.

Screenshot 2023-03-16 at 01.42.52.png

````python
# Python program to
# demonstrate private members
 
# Creating a Base class
class Base:
    def __init__(self):
        self.a = "GeeksforGeeks"
        self.__c = "GeeksforGeeks"
 
# Creating a derived class
class Derived(Base):
    def __init__(self):
 
        # Calling constructor of
        # Base class
        Base.__init__(self)
        print("Calling private member of base class: ")
        print(self.__c)
 
 
# Driver code
obj1 = Base()
print(obj1.a)
 
 
# Uncommenting print(obj1.c) will
# raise an AttributeError
 
# Uncommenting obj2 = Derived() will
# also raise an AtrributeError as
# private member of base class
# is called inside derived class
````

В приведенном выше примере мы создали переменную c как приватный атрибут. Мы не можем получить прямой доступ к этому атрибуту и даже не можем изменить его значение.

## Наследование

Возможность создания новых абстракций на основе существующих.

Наследование - это способность одного класса получать или наследовать свойства от другого класса.
Когда один класс определяется на основе другого, называется **наследованием**. Класс, от которого получены свойства **родительским** или **базовым**, а класс, который получает свойства - **подклассом** родительского или **дочерним классом**.

Преимущества наследования следующие: 
* Оно представляет отношения Is-A. # class ElectricEngine(Engine): # Is-A Engine
* Оно хорошо отражает реальные отношения.
* Оно обеспечивает возможность повторного использования кода. Нам не нужно писать один и тот же код снова и снова. Кроме того, это позволяет нам добавлять дополнительные функции в класс, не изменяя его.
* Наследование является транзитивным по своей природе, что означает, что если класс B наследуется от другого класса A, то все подклассы B автоматически наследуются от класса A.
* Наследование обеспечивает простую, понятную структуру модели. 
* Наследование приводит к меньшим затратам на разработку и сопровождение.

Различные виды наследования. Язык программирования Python реализует как стандартное одиночное наследование.
Single inheritance:

````python
class Mammal(): # Млекопитающие
    className = 'Mammal'

class Dog(Mammal):
    species = 'Canis lupus'

dog = Dog()
dog.className # Mammal
````

Многоуровневое наследование. Multilevel inheritance:
```
class Base(object):
    {body}
class Child(base):
    {body}
class GrandChild(Child):
    {body}
```

````python
# Python program to demonstrate
# multilevel inheritance

# Base class


class Grandfather:

	def __init__(self, grandfathername):
		self.grandfathername = grandfathername

# Intermediate class


class Father(Grandfather):
	def __init__(self, fathername, grandfathername):
		self.fathername = fathername

		# invoking constructor of Grandfather class
		Grandfather.__init__(self, grandfathername)

# Derived class


class Son(Father):
	def __init__(self, sonname, fathername, grandfathername):
		self.sonname = sonname

		# invoking constructor of Father class
		Father.__init__(self, fathername, grandfathername)

	def print_name(self):
		print('Grandfather name :', self.grandfathername)
		print("Father name :", self.fathername)
		print("Son name :", self.sonname)


# Driver code
s1 = Son('Prince', 'Rampal', 'Lal mani')
print(s1.grandfathername)
s1.print_name()
````

Иерархическое наследование. Hierarchical inheritance. Когда из одного базового класса создается более одного производного класса.

````python
# Python program to demonstrate
# Hierarchical inheritance
 
 
# Base class
class Parent:
    def func1(self):
        print("This function is in parent class.")
 
# Derived class1
 
 
class Child1(Parent):
    def func2(self):
        print("This function is in child 1.")
 
# Derivied class2
 
 
class Child2(Parent):
    def func3(self):
        print("This function is in child 2.")
 
 
# Driver's code
object1 = Child1()
object2 = Child2()
object1.func1()
object1.func2()
object2.func1()
object2.func3()
````

Гибридное наследование. Hybrid inheritance. Эта форма сочетает в себе более одной формы наследования. По сути, это смесь более чем одного вида наследования.

Screenshot 2023-04-27 at 01.28.44.png

````python
# Python program to demonstrate
# hybrid inheritance


class School:
	def func1(self):
		print("This function is in school.")


class Student1(School):
	def func2(self):
		print("This function is in student 1. ")


class Student2(School):
	def func3(self):
		print("This function is in student 2.")


class Student3(Student1, School):
	def func4(self):
		print("This function is in student 3.")


# Driver's code
object = Student3()
object.func1()
object.func2()
````

## Полиморфизм

Возможность реализации наследуемых свойств или методов отличающимися способами в рамках множества абстракций.

В зависимости от того, какой класс(подкласс) используется, происходят различные действия, при этом не нужно явно знать, что это за класс(подкласс).

Действие воспроизведения AudioFile может быть очень простым:
```
 audio_file.play()
```
(Пример с аудиофайлом выше, включающий наследование)

````python

from pathlib import Path
from abc import ABC,abstractmethod


class Playable(ABC):
    
    @abstractmethod
    def play(self) -> None:
        ...


class AudioFile(Playable):
    ext: str 

    def __init__(self, filepath: Path) -> None:
        if not filepath.suffix == self.ext:
            raise ValueError("Invalid file format")
        self.filepath = filepath


class MP3File(AudioFile):
    ext = ".mp3"

    def play(self) -> None:
        print(f"playing {self.filepath} as mp3")


class WavFile(AudioFile):
    ext = ".wav"

    def play(self) -> None:
        print(f"playing {self.filepath} as wav")


class OggFile(AudioFile):
    ext = ".ogg"

    def play(self) -> None:
        print(f"playing {self.filepath} as ogg")


class FlacFile(Playable):
    def __init__(self, filepath: Path) -> None:
        if not filepath.suffix == ".flac":
            raise ValueError("Not a .flac file")
        self.filepath = filepath

    def play(self) -> None:
        print(f"playing {self.filepath} as flac")





class Entertainment:
    def __init__(self, play_list: list[Playable]) -> None:
        self.queue = play_list

    def play(self) -> None:
        for song in self.queue:
            song.play()


def fragile() -> None:
    s1 = WavFile(Path("Roundabout.wav"))
    s2 = MP3File(Path("Cans and Brahms.mp3"))
    s3 = OggFile(Path("We Have Heaven.ogg"))
    s4 = FlacFile(Path("South Side of the Sky.flac"))
    side_1 = Entertainment([s1, s2, s3, s4])
    side_1.play()

if __name__ == "__main__":
    fragile()
````

Выражение композиции c помощью uml.

Композиция представлена через линию с залитым ромбом у композиционного класса, указывающую на класс компонента. Сторона композита может выражать кардинальность отношений. Кардинальность указывает на количество или допустимый диапазон экземпляров компонентов, которые будет содержать класс Composite.

На диаграмме выше цифра 1 означает, что класс Composite содержит один объект типа Component. Кардинальность может быть выражена следующим образом:

* Число указывает на количество экземпляров Component, которые содержатся в Composite.
* Символ * указывает на то, что класс Composite может содержать переменное количество экземпляров Component.
* Диапазон 1...4 указывает на то, что класс Composite может содержать определенное количество экземпляров компонентов. Диапазон обозначается минимальным и максимальным количеством экземпляров, или минимальным и большим количеством экземпляров, как в 1...*.
