```python
# учитывая уже стощие на доске фигуры, нахожу свободные безопасные клетки
# передаём функции значения: n-количество столбцов и строк; data_fig-спискок координат уже стоящих фигур; 
# svobod_kletki-список, куда будут добавляться свободные безопастные координаты; 
def F_svobod_kletki(n: int, data_fig: list, svobod_kletki: list):
    print('222222222222',svobod_kletki)
    x=0
    y=0
    while  y<=n:
        while x<=n:            
            koordinata =' '
            print('44444444444',data_fig)
            for i in data_fig: #беру координаты первой поставленной фигуры и проверяю заходит ли координата (x;y) на поле её удара,
                a=i[0] # х для уже заданной фигуры
                b=i[1] # у для уже заданной фигуры
                
                if ((y >= (b-4) or y <= (b+4)) and (x >= (a-4) or x <= (a+4))) and ((abs(x-y) == abs(a-b)) or ((x+y)==(a+b))):
                    koordinata ='занята'
                    break #выходим т.к. в этом случае координата нам не подходит
                else:
                    koordinata ='свободна'
            if koordinata == 'свободна':
                print('3333333333',[x,y])
                svobod_kletki.append([x,y])

            x+=1 #проверяем всю строку
        x=0 
        y+=1 #проверяем все столбцы

    return svobod_kletki

# передаём функции значения: n-количество столбцов и строк; l-количество фигур, которые нужно поставить;
# data_fig-спискок координат уже стоящих фигур; svobod_kletki-список, куда будут добавляться; k - счётчик добавленного в output.txt
# flag - переключатель, чтобы на первом витке функция продолжила работу
def check(n: int,l: int, data_fig: list,svobod_kletki: list, k: int, flag: int):
    print('1111111',svobod_kletki)
    if svobod_kletki == []:
        if flag == 1:
            return k #чтобы узнать было ли хоть что-то выведено
        else:
            flag = 1
    if l==0:
        k+=1 
        #выводим
    else:
        print("xcvbnm,.",data_fig)
        svobod_kletki = F_svobod_kletki(n, data_fig, svobod_kletki)
        d=svobod_kletki.pop(0)
        data_fig.append(d)          
        check(n,l-1,data_fig,svobod_kletki)

    return data_fig

def main():
    t= open("input.txt", "r+")
    data = t.readlines()

    # убираю знаки табуляции, чтобы не мешали
    for i in range (len(data)):
        if data[i][-1] == '\n':  
            data[i] = data[i][:-1]
    print('полученные данные', data)

    # преобразую список для удобства
    del0 = data[0].split() # количество столбцов и строк
    n = int(del0[0]) # количество строк и столбцов
    l = int(del0[1]) # сколько фигур необходимо расставить
    k = int(del0[2]) # количесвто уже расставленных фигур
    del0=data.pop(0) # теперь в data только кооринаты начальных фигур
    print('нач. координаты',data)
    print(n)
    print(l)
    print(k)

    # уберём пробел для удобства между координатами
    data_fig = []
    for i in range (k):
        data_fig.append(data[i].split())
    print('data_fig',data_fig)

    # изменим str на int
    for i in data_fig:
        i[0]=int(i[0])
        i[1]=int(i[1])

    # объявим переменные
    flag=0
    k=0
    x=0 
    y=0
    svobod_kletki=[]
                        
    rez = check(n, l, data_fig,[],k,flag)  
    print(len(rez))
    print(rez)

if __name__=="__main__":
    main()         
        
        

     




    

#в рекурсии количество требуемых фигур точка остановки




```
