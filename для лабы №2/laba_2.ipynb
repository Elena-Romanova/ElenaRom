```python
from time import time
# учитывая уже стощие на доске фигуры, нахожу клетки, находщиеся под боем этих фигур
# передаём функции значения: n-количество столбцов и строк; data_fig-спискок координат уже стоящих фигур; 
def F_occupied_kletki(n: int, data_fig: list) -> list:
    occupied_kletki=[]
    for el in data_fig: #беру координаты первой поставленной фигуры и проверяю заходит ли координата (x;y) на поле её удара
        i=el[0]
        j=el[1]
        occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i-1,j-1],[i-2,j-2],[i-3,j-3],[i-4,j-4]])))
        occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i+1,j+1],[i+2,j+2],[i+3,j+3],[i+4,j+4]])))
        occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i-1,j+1],[i-2,j+2],[i-3,j+3],[i-4,j+4]])))
        occupied_kletki.extend(list(filter(lambda x: 0<=x[0]<=n and 0<=x[1]<=n,[[i+1,j-1],[i+2,j-2],[i+3,j-3],[i+4,j-4]])))
    occupied_kletki.extend(data_fig)  
    
    return occupied_kletki

# передаём функции значения: n-количество столбцов и строк; l-количество фигур, которые нужно поставить;
# data_fig-спискок координат уже стоящих фигур; occupied_kletki-список, куда будут добавляться "занятые" клетки;
# x, y - координаты предыдущей позиции
def check(n: int,l: int, data_fig: list, occupied_kletki: list,f,x,y) -> None:
    global reshenie # переменная, которая сохранит последнее решение, чтобы вывести его на консоль
    if l==0:
        for i in range (len(data_fig)):
            data_fig[i]= tuple(data_fig[i])
            f.write (f" {data_fig[i]}")
        reshenie=data_fig
        f.write("\n")
        return False # чтобы можно было вернуть предыдущие значения переменным во избежании ошибочных результатов
    else:
        # проходимся по воображаемому полю, учитывая предыдущие координаты, дабы избежать повторений
        for i in range (x,n+1): 
            for j in range (n+1): 
                if x==i and j<y:
                    continue
                else:
                    if ([i,j] in occupied_kletki):
                        pass
                    else:
                        data_fig = data_fig + [[i,j]]
                        occupied_kletki_do = occupied_kletki 
                        occupied_kletki = F_occupied_kletki (n, data_fig)
                        if not (check(n,l-1,data_fig,occupied_kletki,f,i,j)):
                            data_fig = data_fig[:-1]
                            occupied_kletki = occupied_kletki_do

reshenie= []
def main():
    m=time()
    t= open("input.txt", "r+")
    data = t.readlines()
    f = open("output.txt", "w")

    # убираю знаки табуляции, чтобы не мешали
    for i in range (len(data)):
        data[i] = data[i].strip()

    # преобразую список для удобства
    del0 = data[0].split() # количество столбцов и строк
    n,l,k = int(del0[0]),int(del0[1]), int(del0[2]) # количество строк и столбцов,сколько фигур необходимо расставить,количесвто уже расставленных фигур
    n = n - 1
    del0=data.pop(0) # теперь в data только кооринаты начальных фигур

    # уберём пробел для удобства между координатами
    data_fig = [data[i].split() for i in range (k)]


    # изменим str на int
    for i in data_fig:
        i[0]=int(i[0])
        i[1]=int(i[1])
    
    # вызовем функцию
    occupied_kletki =[]
    occupied_kletki = F_occupied_kletki (n, data_fig) 

    check(n, l, data_fig, occupied_kletki,f,0,0)
    

    # печатаем на консоль
    if reshenie == []:
        f.write('No solutions')
        print('No solutions')
    else:
        occupied_kletki = F_occupied_kletki (n, reshenie)
        for i in range (len(occupied_kletki)):
            occupied_kletki[i]=tuple(occupied_kletki[i])
        n=n+1
        for i in range (n):
            x=''
            for j in range (n):
                if (i,j) in reshenie:
                    x+='#'
                elif (i,j) in occupied_kletki:
                    x+='*'
                else:
                    x+='0'
            print(' '.join(list(x)))

    # f.close()
    # t.close()
    print('Время: '+str(time()-m))

if __name__=="__main__":
    main()  
```
