```python
# учитывая уже стощие на доске фигуры, нахожу свободные безопасные клетки
# передаём функции значения: n-количество столбцов и строк; data_fig-спискок координат уже стоящих фигур; 
# svobod_kletki-список, куда будут добавляться свободные безопастные координаты; 
def F_svobod_kletki(n: int, data_fig: list):
    i=0
    j=0
    svobod_kletki=[]
    while  i<=n:
        while j<=n:            
            koordinata =' '

            for l in data_fig: #беру координаты первой поставленной фигуры и проверяю заходит ли координата (x;y) на поле её удара,
                a=l[0] # х для уже заданной фигуры
                b=l[1] # у для уже заданной фигуры
                
                if ((b>=j>=(b-4) or b<=j<=(b+4)) and (a>=i>=(a-4) or a<=i<=(a+4)) and ((i-j == a-b) or (i+j==a+b))):
                    koordinata ='занята'
                    break #выходим т.к. в этом случае координата нам не подходит
                else:
                    koordinata ='свободна'
            if koordinata == 'свободна':
                svobod_kletki.append([i,j])

            j+=1 #проверяем всю строку
        j=0 
        i+=1 #проверяем все столбцы
    
    return svobod_kletki

# передаём функции значения: n-количество столбцов и строк; l-количество фигур, которые нужно поставить;
# data_fig-спискок координат уже стоящих фигур; svobod_kletki-список, куда будут добавляться 
# vs_povtor - список, который проверяет на повторение; 
def check(n: int,l: int, data_fig: list, svobod_kletki: list,f,vs_povtor: list):
    global reshenie
    k=0
    if l==0:
        for i in range (len(data_fig)):
                data_fig[i]= tuple(data_fig[i])
        for i in vs_povtor:
            if sorted(data_fig)==sorted(i):
                k=1
        if k == 0:
            for i in range (len(data_fig)):
                f.write (f" {data_fig[i]}")
            reshenie=data_fig
            vs_povtor.append(data_fig)
            f.write("\n")
    else:
        svobod_kletki = F_svobod_kletki (n, data_fig)  #находим все своб/клетки от уже стоящ/постав фигур
        for i in svobod_kletki:        
            check(n,l-1,data_fig + [i], svobod_kletki[1:],f,vs_povtor)

    
reshenie= []
def main():
    t= open("input.txt", "r+")
    data = t.readlines()
    f = open("output.txt", "w")

    # убираю знаки табуляции, чтобы не мешали
    for i in range (len(data)):
        if data[i][-1] == '\n':  
            data[i] = data[i][:-1]

    # преобразую список для удобства
    del0 = data[0].split() # количество столбцов и строк
    n = int(del0[0]) # количество строк и столбцов
    n = n - 1
    l = int(del0[1]) # сколько фигур необходимо расставить
    k = int(del0[2]) # количесвто уже расставленных фигур
    del0=data.pop(0) # теперь в data только кооринаты начальных фигур

    # уберём пробел для удобства между координатами
    data_fig = []
    for i in range (k):
        data_fig.append(data[i].split())

    # изменим str на int
    for i in data_fig:
        i[0]=int(i[0])
        i[1]=int(i[1])
    
    # вызов функции
    svobod_kletki = [] 
    vs_povtor = []     
    check(n, l, data_fig, svobod_kletki,f,vs_povtor)  

    # печатаем на консоль
    svobod_kletki = F_svobod_kletki (n, reshenie)
    for i in range (len(svobod_kletki)):
        svobod_kletki[i]=tuple(svobod_kletki[i])
    n=n+1
    for i in range (n):
        x=''
        for j in range (n):
            if (i,j) in reshenie:
                x+='#'
            elif (i,j) in svobod_kletki:
                x+='0'
            else:
                x+='*'
        print(' '.join(list(x)))

    f.close()
    t.close()

if __name__=="__main__":
    main()  
```
