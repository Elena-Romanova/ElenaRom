object - может быть пустым/заполненным кортежем, списком, набором, словарем и т.д. или любым объектом, определенным пользователем

Каждый модуль создает свое собственное глобальное пространство имен (global namespace). Эти различные пространства имен изолированы. Следовательно, одно и то же имя, которое может существовать в разных модулях, не сталкивается. Модули могут иметь различные функции и классы. Они имеют разное время жизни. По мере выполнения программы Python создает пространства имен по мере необходимости и удаляет их, когда они больше не нужны. Как правило, в каждый момент времени существует множество пространств имен.

Когда происходит поиск ссылки внутри функции, то имя ищется в локальном пространстве имен, затем в глобальном пространстве имен и, наконец, во встроенном пространстве имен.
Если функция находится внутри другой функции, новая область видимости вложена в локальную область видимости.


Если мы пишем программу, например, в файле myprog.py, то автоматически создается пространство имен с именем нашего файла. Если мы объявляем переменные, то в этом пространстве имен появляются соответствующие переменные. Все переменные, которые находятся в этом пространстве имен, называются гобальными.

Example 1: Scope and Namespace in Python

````python
# global_var is in the global namespace
global_var = 10

def outer_function(): #  the enclosing function
    #  outer_var is in the local namespace 
    outer_var = 20

    def inner_function(): # the enclosed function
        #  inner_var is in the nested local namespace 
        inner_var = 30

        print(inner_var)

    print(outer_var)

    inner_function()

# print the value of the global variable
print(global_var)

# call the outer function and print local and nested local variables
outer_function()
````

В приведенном выше примере есть три отдельных пространства имен: глобальное пространство имен, локальное пространство имен во внешней функции и локальное пространство имен во внутренней функции.

````python
x = 'global'

def f():
    x = 'enclosing'

    def g():
        x = 'local'
        print(x)
    print(x)
    g()
f()
````

Example 2: Use of global Keyword in Python

````python
# define global variable 
global_var = 10

def my_function():
    # define local variable
    local_var = 20

    # modify global variable value 
    global global_var
    global_var = 30

# print global variable value
print(global_var)

# call the function and modify the global variable
my_function()

# print the modified value of the global variable
print(global_var)
````

Здесь, когда вызывается функция, ключевое слово global используется для указания на то, что global_var является глобальной переменной, и ее значение изменяется на 30.

### Python Namespace Dictionaries

### The locals() function
Функция locals()

Python также предоставляет соответствующую встроенную функцию, называемую locals(). Она похожа на globals(), но вместо этого получает доступ к объектам в локальном пространстве имен:
```
>>> def f(x, y):
...     s = 'foo'
...     print(locals())
...

>>> f(10, 0.5)
{'s': 'foo', 'y': 0.5, 'x': 10}
```

При вызове внутри f() функция locals() возвращает словарь, представляющий локальное пространство имен функции. Обратите внимание, что, помимо локально определенной переменной s, локальное пространство имен включает параметры функции x и y, поскольку они также являются локальными для f().

Если вы вызываете locals() вне функции в основной программе, то она ведет себя так же, как и globals().

### Deep Dive: A Subtle Difference Between globals() and locals()

Глубокое погружение: Тонкое различие между globals() и locals()

Между globals() и locals() есть одно небольшое различие, о котором полезно знать.

globals() возвращает фактическую ссылку на словарь, содержащий глобальное пространство имен. Это означает, что если вы вызовете globals(), сохраните возвращаемое значение и впоследствии определите дополнительные переменные, то эти новые переменные появятся в словаре, на который указывает сохраненное возвращаемое значение:
```
>>> g = globals()
 2>>> g
 3{'__name__': '__main__', '__doc__': None, '__package__': None,
 4'__loader__': <class '_frozen_importlib.BuiltinImporter'>, '__spec__': None,
 5'__annotations__': {}, '__builtins__': <module 'builtins' (built-in)>,
 6'g': {...}}
 7
 8>>> x = 'foo'
 9>>> y = 29
10>>> g
11{'__name__': '__main__', '__doc__': None, '__package__': None,
12'__loader__': <class '_frozen_importlib.BuiltinImporter'>, '__spec__': None,
13'__annotations__': {}, '__builtins__': <module 'builtins' (built-in)>,
14'g': {...}, 'x': 'foo', 'y': 29}
```


Здесь g - это ссылка на словарь глобального пространства имен. После операторов присваивания в строках 8 и 9, x и y появляются в словаре, на который указывает g.

locals(), с другой стороны, возвращает словарь, который является текущей копией локального пространства имен, а не ссылкой на него. Дальнейшие добавления в локальное пространство имен не повлияют на предыдущее возвращаемое значение функции locals(), пока вы не вызовете ее снова. Также вы не можете изменять объекты в реальном локальном пространстве имен, используя возвращаемое значение из locals():

В этом примере loc указывает на возвращаемое значение из locals(), которое является копией локального пространства имен. Оператор x = 20 в строке 6 добавляет x в локальное пространство имен, но не в копию, на которую указывает loc. Аналогично, оператор в строке 9 изменяет значение ключа 's' в копии, на которую указывает loc, но это не влияет на значение s в реальном локальном пространстве имен.

Это тонкое различие, но оно может вызвать у вас проблемы, если вы его не запомните.

### The global Declaration

Если же мы хотим поменять именно глобальную переменную, то в функции следует вначале это явно сказать:

Оператор global x указывает, что во время выполнения f() ссылки на имя x будут ссылаться на x, который находится в глобальном пространстве имен. Это означает, что присваивание x = 40 не создает новую ссылку. Вместо этого оно присваивает новое значение x в глобальной области видимости:


Нет особых причин делать это таким образом, поскольку глобальное объявление, возможно, делает намерение более ясным. Но это дает еще одну иллюстрацию того, как работает globals().

Однако, здесь следует быть аккуратным, так как если в функции уже была создана локальная a, то конструкция global приведет к ошибке:

### The nonlocal Declaration

Глобальное объявление позволяет функции получать доступ и изменять объект в "enclosed" области видимости. А что если вложенной функции нужно изменить объект в внешней области видимости? Имена, указанные после ключевого слова nonlocal, относятся к переменным в ближайшей внешней области видимости:

```Py
>>> def f():
 2...     x = 20
 3...
 4...     def g():
 5...         nonlocal x
 6...         x = 40
 7...
 8...     g()
 9...     print(x)
10...
11
12>>> f()
1340
```

После нелокального оператора x в строке 5, когда g() ссылается на x, она ссылается на x в ближайшей объемлющей области видимости, определение которой находится в f() в строке 2:

Screenshot 2023-03-23 at 03.26.48.png

Оператор print() в конце f() в строке 9 подтверждает, что вызов g() изменил значение x в объемлющей области видимости на 40.

````python
x = 0
def outer():
    x = 1
    def inner():
        nonlocal x
        x = 2
        print("inner:", x)
 
    inner()
    print("outer:", x)
 
outer()
print("global:", x)
````

Несмотря на то, что Python предоставляет ключевые слова global и nonlocal, использовать их не всегда целесообразно.

Как и многое другое, это в некоторой степени вопрос стиля и предпочтений. Бывают случаи, когда разумное использование модификации глобальных переменных может уменьшить сложность программы.

В целом, модификация переменных за пределами локальной области видимости обычно не нужна. Почти всегда есть лучший способ, обычно с возвращаемыми значениями функций.

# Closures
Замыкания

Замыкания в Python - это вложенная функция, которая позволяет нам получить доступ к переменным внешней функции даже после того, как внешняя функция закрыта.

Прежде чем мы узнаем о замыкании, давайте сначала рассмотрим концепцию вложенных функций (nested functions) в Python.

### Nested function in Python

````python
def greet(name):
    # inner function
    def display_name():
        print("Hi", name)
    
    # call inner function
    display_name()

# call outer function
greet("John")  

# Output: Hi John
````

В приведенном выше примере мы определили функцию display_name() внутри функции greet().

Здесь display_name() является вложенной функцией. Вложенная функция работает аналогично обычной функции. Она выполняется, когда display_name() вызывается внутри функции greet().

````python
def greet():
    # variable defined outside the inner function
    name = "John"
    def display_name():
        print("Hi", name)
        # return a nested function
        
    return display_name # Возвращаем ссылку на функцию, а не выполняем ее
````

### Python Closures

Замыкание (closure) — это вложенная функция, которая ссылается на одну или более переменных из объемлющей (enclosing) области видимости.


Как мы уже обсуждали, замыкание - это вложенная функция, которая помогает нам получить доступ к переменным внешней функции даже после того, как внешняя функция закрыта. Например,

````python
def greet():
    # variable defined outside the inner function
    name = "John"
    def display_name():
        print("Hi", name)
        # return a nested function

    return display_name # Возвращаем ссылку на функцию, а не выполняем ее

# call the outer function
message = greet() # присваиваем возвращаемое значение функции greet() переменной message. (Сохнаняем ссылку на внутреннюю функцию)
# call the inner function
# Вызываем внутреннюю функцию, через ссылку message
message() # Когда выполняется message(), функция greet() уже завершила выполнение. 

# Output: Hi John
````

Когда у нас имеется глобальная ссылка message на локальное окружение функции display_name, то это окружение продолжает существовать, т.е. оно не удаляется сборщиком мусора. (Так как глобальная переменная message ссылается на это окружение). 
А вместе с локальным окружением display_name продолжают существовать и все остальные внешние окружения, которые с ним связаны. Потому что у каждого локально окружения есть неявная ссылка на внешнее окружение. В данном случае у функции display_name внешнее окружение это функция greet, соответсвтенно у функции greet идет ссылка на глобальное окружение нашей программы (имя модуля). Все эти окружение не пропадут, пока существует ссылка message.
Блакодаря этому, функция display_name может спокойно обращаться к переменной внешнего окружения name и уже с ней оперировать.

Замыкание в том смысле, что мы "держим" эти окружения "по цепочке".

У нас есть переменная message, которая пренадлежит глобальной области видимости.

Окружение функции display_name ссылается на внешнее окружение функции greet.
Окружение функции greet ссылается на глобальное окружение.

Т.е. у нас получается своеобразная замкнутая цепочка ссылок.


```

```

При каждом вфзове функции greet будет создаваться свое локальное окружение этой функции:

````python
def greet(name):
    # variable defined outside the inner function
    def display_name():
        print("Hi", name)
        # return a nested function

    return display_name # Возвращаем ссылку на функцию, а не выполняем ее

# call the outer function
message = greet("Jonh") # создается локальное окружение
message1 = greet("Hue") # создается локальное окружение

message()
message1()

message
````

### Example: Counter
Функция-счетчик. Увеличиваем значение на 1 при каждом вызове

Через замыкание можно создавать независимые счетчики.

````python
def counter(start=0):
    def step():
        nonlocal start
        start += 1
        return start
    return step

c1 = counter(10)
c2 = counter()
print(c1(),c2())
print(c1(),c2())
print(c1(),c2())
print(c1(),c2())
````

Как мы видим, оба счетчика работают независимо друг от друга.

### Pass Function as Argument
Передача функции в качестве аргумента

В Python мы можем передать функцию в качестве аргумента другой функции. Например,

````python
def add(x, y):
    return x + y

def calculate(func, x, y): # <---
    return func(x, y)

result = calculate(add, 4, 6)
print(result)  # prints 10
````

## Python Decorators
Декораторы Python

Как упоминалось ранее, декоратор Python - это функция, которая принимает в качестве аргумента функцию и возвращает ее, добавляя некоторую функциональность.

Фактически, любой объект, реализующий специальный метод ```__call__()```, называется вызываемым. Таким образом, в самом базовом смысле декоратор - это вызываемый объект, который возвращает вызываемый объект.

По сути, декоратор принимает функцию в качестве аргумента, добавляет некоторую функциональность и возвращает ее.

Вы определяете логику или функциональность отдельно от функций, которые будут пользоваться этой функциональностью. Возможно, эту логику разрабатывает для вас другая команда.

````python
def make_pretty(func):
    def inner():
        print("I got decorated")
        func()
    return inner # Возвращаем ссылку на внутр-ю ф-юю inner.


def ordinary():
    print("I am ordinary")

ordinary()
# Output: I am ordinary
````

Здесь мы объявили две функции:

* ordinary(), которая печатает "I am ordinary". # Я обычный
* make_pretty(), которая принимает функцию в качестве аргумента и имеет вложенную функцию inner(), и возвращает ее в return.

Мы вызываем функцию ordinary() обычным способом, поэтому получаем на выходе "I am ordinary". Теперь давайте вызовем ее с помощью функции-декоратора.

````python
def make_pretty(func):
    # define the inner function 
    def inner():
        # add some additional behavior to decorated function
        # Можем выполнить действия до вызова функции func
        print("I got decorated")
        # call original function
        func()
        print("I got decorated")
        # Можем выполнить действия после вызова функции func
        
    # return the inner function
    return inner

# define ordinary function
def ordinary():
    print("I am ordinary")
    
# decorate the ordinary function
#decorated_func = make_pretty(ordinary) 
# Часто, чтобы не создавать новых переменных пишут:
ordinary = make_pretty(ordinary)

# call the decorated function
#decorated_func()

ordinary()
````

### @ Symbol With Decorator
@ символ.

Вместо того чтобы присваивать вызов функции переменной, Python предлагает гораздо более элегантный способ достижения этой функциональности с помощью символа @. 
Например,

````python
def make_pretty(func):

    def inner():
        print("I got decorated")
        func()
    return inner

@make_pretty # Декоратор, который хотим применить к функции ordinary
def ordinary():
    print("I am ordinary")

ordinary()  
````

Здесь функция ordinary() "декорирована" декоратором make_pretty() с использованием синтаксиса @make_pretty, что эквивалентно вызову 
```
ordinary = make_pretty(ordinary).
```

Это также называют синтаксическим сахаром.

Синтаксический сахар (syntactic sugar) — это любой элемент синтаксиса языка программирования, который даёт программисту альтернативный способ записи другой, уже имеющейся в языке синтаксической конструкции, и при этом является более удобным, или более кратким, или похожим на другой распространённый способ записи, или помогает писать программы в хорошем стиле.

```
@gfg_decorator
def hello_decorator():
    print("Gfg")

'''Above code is equivalent to -

def hello_decorator():
    print("Gfg")
    
hello_decorator = gfg_decorator(hello_decorator)'''
```

````python
# объявляем decorator
def hello_decorator(func):

	# inner1 - это функция-обертка, в которой вызывается аргумент
	
	# внутренняя функция может обращаться к внешним локальным функциям, как в данном случае "func"
	def inner1():
		print("Hello, this is before function execution")

		# вызов фактической функции теперь внутри функции-обертки.
		func()

		print("This is after function execution")
		
	return inner1


# определение функции, которая будет вызываться внутри обертки
def function_to_be_used():
	print("This is inside the function !!")


# передаем 'function_to_be_used' внутрь декоратора для управления ee поведением
function_to_be_used = hello_decorator(function_to_be_used)


# Вызываем функцию
function_to_be_used() # ссылка на wrapper/inner1
````

### Example

````python
def divide(a, b):
    return a/b

divide(10,0)
````

Эта функция имеет два параметра, a и b. Мы знаем, что она выдаст ошибку, если мы передадим b равным 0.

Теперь давайте сделаем декоратор для проверки этого случая, который приведет к ошибке.

````python
def smart_divide(func):
    def inner(a, b):
        print("I am going to divide", a, "and", b)
        if b == 0:
            print("Whoops! cannot divide")
            return

        return func(a, b)
    return inner

@smart_divide
def divide(a, b):
    print(a/b)

divide(2,5)

divide(2,0)
````

Здесь, когда мы вызываем функцию divide() с аргументами (2,5), вместо нее вызывается функция inner(), определенная в декораторе smart_divide().

Эта функция inner() вызывает исходную функцию divide() с аргументами 2 и 5 и возвращает результат, равный 0,4.

Аналогично, когда мы вызываем функцию divide() с аргументами (2,0), функция inner() проверяет, что b равно 0, и выводит сообщение об ошибке, после чего возвращает None.

### Chaining Decorators in Python

Цепочка декораторов в Python

В Python можно объединить несколько декораторов в цепочку.

Для создания цепочки декораторов в Python мы можем применить несколько декораторов к одной функции, расположив их один за другим.

````python
def star(func):
    def inner(*args, **kwargs):
        print("*" * 15)
        func(*args, **kwargs)
        print("*" * 15)
    return inner


def percent(func):
    def inner(*args, **kwargs):
        print("%" * 15)
        func(*args, **kwargs)
        print("%" * 15)
    return inner


@star
@percent
def printer(msg):
    print(msg)

printer("Hello")
````

Приведенный выше синтаксис,
```Py
@star
@percent
def printer(msg):
    print(msg)
```
эквивалентен
```Py
def printer(msg):
    print(msg)
printer = star(percent(printer))
```
Порядок, в котором мы выстраиваем цепочки декораторов, имеет значение. Если бы мы изменили порядок на противоположный,
```Py
@percent
@star
def printer(msg):
    print(msg)
```
Вывод был бы следующим:

```

Hello
```

# Example. Runtime
С помощью декоратора, например, можем проверить время выполнения различных функций.

````python
import time

def testTime(fn):
    def wrapper(*args, **kwargs):
        st = time.time()
        res = fn(*args, **kwargs)
        dt = time.time() - st
        print(f"Время работы: {dt} сек")
        return res
    return wrapper

@testTime
def JustForLoop(x=0):
    str = ''
    for i in range(x):
        str += "*"
    return "some res"


JustForLoop(100000)
res = JustForLoop(100000000)
print(res)
````
